```python
#import pandas, load in as shorthand so we can use it easier- don't have to write pandas every time
import pandas as pd
```

```python
#load in data from csv files
df_test = pd.read_csv('test_data.csv')
df_students = pd.read_csv('students.csv')
df_sites = pd.read_csv('sites.csv')
```

```python
df_test.head()
```

```python
df_students.head()
with pd.option_context('display.max_rows', None):
    print(df_students)
```

```python
df_sites.head()
```

```python
#merge dataframes of students and sites together on 'site_id', preserving all unmatched values from both dataframes
df_students_sites = pd.merge(df_students, df_sites, how='outer', on='site_id')

#rename site_name to school_name for consistency with test data
df_students_sites.rename(columns={'site_name': 'school_name'}, inplace=True)

#merge dataframe df_student_sites, preserving all unmatched values from both dataframes
df_merged = pd.merge(df_students_sites, df_test, how='outer')
```

```python
#Anne - (1) match site name site_id, (2) match both to to school name, (3) match all to test date
```

```python
#print all rows and columns of merged dataframe
with pd.option_context('display.max_rows', None):
    print(df_merged)
```

```python
#rename 'score' to 'test_score', as specified in the instructions
df_merged.rename(columns={'score': 'test_score'}, inplace=True)
```

```python
# Assuming your merged data is called matched_data
print('Match counts: ')
print(df_merged.notna().value_counts() )
```

```python
with pd.option_context('display.max_rows', None):
    print(df_merged)
```

```python
#Make CSV with specified columns and no index
df_merged.to_csv('matched_student_data.csv', columns=['student_id', 'test_date', 'test_score'], index=False)
```
